```python
def confusion_m(threshold):
    ridge_regression_estimator = rr.RidgeRegression(l2reg=0.01)
    ridge_regression_estimator.fit(X_train, y_train)
    our_coefs = ridge_regression_estimator.w_
    new_coefs = np.where(np.absolute(our_coefs) < threshold, 0, 1)
    coefs = np.where(coefs_true ==0,0,1)
    print("When threshold =",threshold,"confusion_matrix is:\n",confusion_matrix(coefs, new_coefs))
confusion_m(10**-6)
confusion_m(10**-3)
confusion_m(10**-1)
```

## 3.1.2
### 3.2.1.function that computes the Lasso solution for a given lambda

```python
def compute_square_loss(X, y, w, lambda_reg):
    loss = 0 #Initialize the average square loss
    num_instances = X.shape[0]
    num_features =  X.shape[1]
    factor = np.matmul(X,w) - y
    loss = np.matmul(factor.transpose(),factor).mean()  
    return loss

#coordinate descent
def lasso_solution(X, y, lambda_reg, use_random = False, w=None):
    num_instances,num_features = X.shape[0], X.shape[1]
    
    if w is None:
        w = np.zeros(num_features)
    loss_hist = []
    loss1 = compute_square_loss(X,y,w,lambda_reg)
    loss_hist.append(loss1)
    epoch = 0
    
    while(epoch<1000):
        idx = np.arange(num_features)
        if use_random:
            np.random.shuffle(idx)
        for j in idx:
            square = np.matmul(X[:,j].transpose(),X[:,j])
            a = 2 * square
            c = 2 * (np.matmul(X[:,j].transpose(),y) - np.matmul(np.matmul(X[:,j].transpose(),X),w) + w[j] * square)
            if c < -lambda_reg:
                w[j] = (c+lambda_reg)/a
            elif c > lambda_reg:
                w[j] = (c-lambda_reg)/a
            else:
                w[j] = 0
        loss = compute_square_loss(X,y,w,lambda_reg) + lambda_reg*sum(abs(w))
        loss_hist.append(loss)
        if abs(loss1 - loss)> 10**(-8):
            loss1 = loss
        else:
            break
        epoch += 1
    return w, loss_hist
```

### 3.2.2.compare performance of cyclic coordinate descent to randomized coordinate descent

```python
w_c1,loss_hist_c1 = lasso_solution(X_train, y_train, lambda_reg=1,use_random = False)
w_c2,loss_hist_c2 = lasso_solution(X_train, y_train, lambda_reg=0.1,use_random = False)
w_c3,loss_hist_c3 = lasso_solution(X_train, y_train, lambda_reg=0.01,use_random = False)
w_r1,loss_hist_r1 = lasso_solution(X_train, y_train, lambda_reg=1,use_random = True)
w_r2,loss_hist_r2 = lasso_solution(X_train, y_train, lambda_reg=0.1,use_random = True)
w_r3,loss_hist_r3 = lasso_solution(X_train, y_train, lambda_reg=0.01,use_random = True)
```

```python
fig = plt.gcf()
fig.set_size_inches(12, 4)

plt.subplot(1,2,1)
#figure1
plt.plot(loss_hist_c1, label = 'cyclic coordinate descent: lambda = 1')
plt.plot(loss_hist_c2, label = 'cyclic coordinate descent: lambda = 0.1')
plt.plot(loss_hist_c3, label = 'cyclic coordinate descent: lambda = 0.01')
plt.plot(loss_hist_r1, label = 'randomized coordinate descent: lambda = 1')
plt.plot(loss_hist_r2, label = 'randomized coordinate descent: lambda = 0.1')
plt.plot(loss_hist_r3, label = 'randomized coordinate descent: lambda = 0.01')
plt.xlabel("Epochs")
plt.ylabel("Total Loss_Training")
plt.ylim(0,100)
plt.legend()

plt.subplot(1,2,2)
#figure 2
lambda_list = np.linspace(-3,1,num = 10)
loss_train_list_c = []
loss_val_list_c = []
loss_train_list_r = []
loss_val_list_r = []
for lambda_reg in lambda_list:
    lambda_reg = 10**(lambda_reg)
    #cyclic
    w_c, loss_hist_c = lasso_solution(X_train,y_train,lambda_reg,use_random=False)
    loss_train_c = loss_hist_c[-1]/X_train.shape[0]
    loss_train_list_c.append(loss_train_c)
    loss_val_c = compute_square_loss(X_val, y_val, w_c,lambda_reg)/X_val.shape[0] 
    loss_val_list_c.append(loss_val_c)
    #randomized
    w_r, loss_hist_r = lasso_solution(X_train,y_train,lambda_reg,use_random=True)
    loss_train_r = loss_hist_r[-1]/X_train.shape[0]
    loss_train_list_r.append(loss_train_r)
    loss_val_r = compute_square_loss(X_val, y_val, w_r,lambda_reg)/X_val.shape[0] 
    loss_val_list_r.append(loss_val_r)
plt.plot(lambda_list,loss_val_list_c, label = 'average square loss_validation_cyclic')
plt.plot(lambda_list,loss_train_list_c, label = 'average square loss_training_cyclic')
plt.plot(lambda_list,loss_val_list_r, label = 'average square loss_validation_randomized')
plt.plot(lambda_list,loss_train_list_r, label = 'average square loss_training_randomized')
plt.xlabel('log($\lambda$)')
plt.ylabel("Average Square Loss")
plt.legend()
plt.show()
```

__Answer:__ <br> As is shown above, randomized coordinate descent, compared to cyclic coordinate descent, is much faster to reach a point that’s close to the minimizer, also it makes lower average square loss for the validation set. So randomized coordinate descent is better in both converge speed and performance.

### 3.2.3.compare solutions attained for starting at 0 VS starting at ridge regression solution

```python
w_ridge = ridge_regression_estimator.w_
w_ridge_r3,loss_hist_ridge_c3 = lasso_solution(X_train,y_train,lambda_reg=0.01,use_random=False,w = w_ridge)
w_ridge_r2,loss_hist_ridge_c2 = lasso_solution(X_train,y_train,lambda_reg=0.1,use_random=False,w = w_ridge)
w_ridge_r1,loss_hist_ridge_c1 = lasso_solution(X_train,y_train,lambda_reg=1,use_random=False,w = w_ridge)
w_ridge_r3,loss_hist_ridge_r3 = lasso_solution(X_train,y_train,lambda_reg=0.01,use_random=True,w = w_ridge)
w_ridge_r2,loss_hist_ridge_r2 = lasso_solution(X_train,y_train,lambda_reg=0.1,use_random=True,w = w_ridge)
w_ridge_r1,loss_hist_ridge_r1 = lasso_solution(X_train,y_train,lambda_reg=1,use_random=True,w = w_ridge)
```

```python
fig = plt.gcf()
fig.set_size_inches(12, 4)

plt.subplot(1,2,1)
plt.plot(loss_hist_c3, label = 'cyclic: $\lambda$ = 0.01,starting at 0')
plt.plot(loss_hist_c2, label = 'cyclic: $\lambda$ = 0.1,starting at 0')
plt.plot(loss_hist_c1, label = 'cyclic: $\lambda$ = 1,starting at 0')
plt.plot(loss_hist_ridge_c3, label = 'cyclic: $\lambda$ = 0.01,starting at ridge regression solution')
plt.plot(loss_hist_ridge_c2, label = 'cyclic: $\lambda$ = 0.1,starting at ridge regression solution')
plt.plot(loss_hist_ridge_c1, label = 'cyclic: $\lambda$ = 1,starting at ridge regression solution')
plt.ylabel("Total loss_Training_cyclic")
plt.xlabel("Epochs")
plt.ylim(0,50)
plt.legend()

plt.subplot(1,2,2)
plt.plot(loss_hist_r3, label = 'randomized: $\lambda$ = 0.01,starting at 0')
plt.plot(loss_hist_r2, label = 'randomized: $\lambda$ = 0.1,starting at 0')
plt.plot(loss_hist_r1, label = 'randomized: $\lambda$ = 1,starting at 0')
plt.plot(loss_hist_ridge_r3, label = 'randomized: $\lambda$ = 0.01,starting at ridge regression solution')
plt.plot(loss_hist_ridge_r2, label = 'randomized: $\lambda$ = 0.1,starting at ridge regression solution')
plt.plot(loss_hist_ridge_r1, label = 'randomized: $\lambda$ = 1,starting at ridge regression solution')
plt.ylabel("Total loss_Training_randomized")
plt.xlabel("Epochs")

plt.ylim(0,50)
plt.legend()
plt.show()
```

```python
pred_fns = []
x = np.sort(np.concatenate([np.arange(0,1,.001), x_train]))

name = "Target Parameter Values (i.e. Bayes Optimal)"
pred_fns.append({"name":name, "coefs":coefs_true, "preds": target_fn(x) })

l2regs = [1.438450]
X = featurize(x)
for l2reg in l2regs:
     
    w, loss_hist = lasso_solution(X_train,y_train,l2reg,use_random=True,w = w_ridge)
    y_pred = np.matmul(X,w)
    name = "Ridge with L1Reg="+str(l2reg)
    pred_fns.append({"name":name,
                    "coefs":w,
                    "preds":y_pred})
l2regs = [0.01, 1]
for l2reg in l2regs:
    ridge_regression_estimator = rr.RidgeRegression(l2reg=l2reg)
    ridge_regression_estimator.fit(X_train, y_train)
    name = "Ridge with L2Reg="+str(l2reg)
    pred_fns.append({"name":name,
                    "coefs":ridge_regression_estimator.w_,
                    "preds": ridge_regression_estimator.predict(X) })


f = rr.plot_prediction_functions(x, pred_fns, x_train, y_train, legend_loc="best")
f.show()
```

```python
f = rr.compare_parameter_vectors(pred_fns)
f.show()
```

__Answer:__ <br>As shown in the bar charts, we can find that while L2 loss creates nice smooth estimator but it is never sparse(almost all parameters are nonzero). Compared to ridge regression, lasso's estimators are sparse. Lasso not only punish high values of the coefficients but actually set them to zero if they are not relevant.
<br>The best lasso model I found is when $\lambda = 1.438450$, using randomized coordinate descent and starting at ridgee regression solution, with its average loss of validation set as 0.122926 .

### 3.1.4 homotopy

```python
df = pd.DataFrame(columns=['params','average_loss_validation'])

lambda_max = 2 * np.linalg.norm(np.matmul(X_train.transpose(),y_train),ord = np.inf)
num_instances = X_val.shape[0]
w, _ = lasso_solution(X_train, y_train, lambda_max,w = None)
loss_val = compute_square_loss(X_val, y_val, w, lambda_max)/num_instances
df.loc[0] = [lambda_max,loss_val]
idx = 1
for i in range(1,30):
    lambda_reg = lambda_max * (0.8**i)
    w, _ = lasso_solution(X_train, y_train, lambda_reg,w = w)
    loss_val = compute_square_loss(X_val, y_val, w, lambda_reg)/num_instances
    df.loc[idx] = [lambda_reg,loss_val]
    idx += 1
   
```

```python
fig, ax = plt.subplots()
ax.semilogx(df["params"], df["average_loss_validation"])
ax.grid()
ax.set_title("Validation Performance vs L1 Regularization")
ax.set_xlabel("L1-Penalty Regularization Parameter")
ax.set_ylabel("Average Squared loss")
fig.show() 
```
